# Comprehensive List of Feature Representations in AI and Math

| **Category**              | **Type of Representation**          | **Description**                                                         | **Common Usage**                                                   |
|----------------------------|-------------------------------------|-------------------------------------------------------------------------|--------------------------------------------------------------------|
| Linear Spaces              | Vector Space                        | Features are represented as vectors (e.g., ℝⁿ), obeying linear algebra rules | Most traditional ML (SVM, logistic regression, deep learning embeddings) |
| Linear Spaces              | Matrix Representation               | Features as structured matrices (2D arrays)                             | Images, tabular data, signal processing                            |
| Linear Spaces              | Tensor Space                        | Multi-dimensional generalization of matrices                            | Deep learning (PyTorch, TensorFlow tensors)                        |
| Probabilistic Spaces       | Probability Distributions           | Features represented as distributions (Gaussian, Bernoulli, Multinomial) | Bayesian models, VAEs, generative models                           |
| Probabilistic Spaces       | Statistical Moments                 | Mean, variance, skewness, kurtosis as feature descriptors               | Feature engineering, generative statistics                         |
| Geometric Spaces           | Euclidean Space                     | Standard flat-space representation (ordinary distances)                 | Most ML, CNNs, clustering (KMeans)                                 |
| Geometric Spaces           | Riemannian Manifolds                | Curved spaces, non-Euclidean geometry                                   | Pose estimation, diffusion models, hyperbolic networks             |
| Geometric Spaces           | Hyperbolic Space                    | Representations where hierarchical structures are naturally encoded     | Knowledge graphs, tree embeddings                                  |
| Topological Spaces         | Topology-Invariant Features         | Focus on connectivity, not distances (e.g., persistent homology)        | Topological data analysis, time-series analysis                    |
| Graph-Based Spaces         | Graph Structures (Nodes + Edges)    | Features embedded in graph form, relations matter                       | GNNs, molecule learning, social network analysis                   |
| Latent Spaces              | Latent Embedding Space              | Low-dimensional hidden representation learned by the model              | Autoencoders, VAEs, GANs                                           |
| Latent Spaces              | Feature Manifolds                   | Assume data lies on a lower-dimensional manifold inside high-d space    | Manifold learning (Isomap, LLE, t-SNE)                             |
| Frequency Domain           | Fourier / Wavelet Transforms        | Features transformed into frequency components                         | Signal processing, audio recognition, some CNN variants            |
| Algebraic Structures       | Group Representations               | Using algebraic groups (rotation, translation symmetries) to encode invariances | Equivariant neural networks, physics-informed models               |
| Logical Spaces             | Symbolic Representations            | Logical symbols, relations, rules as features                          | Symbolic AI, knowledge reasoning systems                           |
| Relational Representations | Set or Multi-Relational Representations | Features are sets, relations among sets are learned                     | Relational learning, relational reinforcement learning             |
| Attention-Based Spaces     | Attention Weights as Representations | Features weighted dynamically based on their relevance                  | Transformers, attention models, sequence modeling                  |
| Complex & Quaternion Spaces| Complex-Valued Representations       | Features are complex numbers or quaternions (4D)                        | Quantum ML, signal processing, rotation-invariant models           |
| Energy-Based Spaces        | Energy Functions                    | Representations are modeled through energy landscapes                   | Energy-based models (EBMs), Hopfield networks                      |
| Metric Learning Spaces     | Distance-Based Embeddings           | Representations optimized to preserve pairwise distances                | Siamese networks, triplet loss embeddings                          |
| Density Spaces             | Density Functions                   | Representing features through probability density (PDF) functions       | Normalizing flows, score-based generative models                   |


```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy.stats import norm

# Create figure
fig = plt.figure(figsize=(18, 18))

# ---------- Linear Spaces ----------
# Vector space
ax1 = fig.add_subplot(5, 4, 1)
v1 = np.array([2, 1])
v2 = np.array([1, 2])
ax1.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='r')
ax1.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='b')
ax1.set_xlim(-1, 3)
ax1.set_ylim(-1, 3)
ax1.set_title("Vector Space")

# Matrix representation (image-like)
ax2 = fig.add_subplot(5, 4, 2)
mat = np.arange(16).reshape(4, 4)
ax2.imshow(mat, cmap="viridis")
ax2.set_title("Matrix Representation")

# Tensor space (3D cube slice)
ax3 = fig.add_subplot(5, 4, 3, projection='3d')
tensor = np.random.rand(3, 3, 3)
ax3.scatter(*np.indices((3, 3, 3)).reshape(3, -1), c=tensor.flatten(), cmap="plasma")
ax3.set_title("Tensor Space")

# ---------- Probabilistic Spaces ----------
# Probability distribution
ax4 = fig.add_subplot(5, 4, 4)
x = np.linspace(-4, 4, 100)
ax4.plot(x, norm.pdf(x, 0, 1), label="Gaussian N(0,1)")
ax4.legend()
ax4.set_title("Probability Distribution")

# ---------- Geometric Spaces ----------
# Euclidean (points in 2D)
ax5 = fig.add_subplot(5, 4, 5)
pts = np.random.randn(100, 2)
ax5.scatter(pts[:, 0], pts[:, 1], alpha=0.6)
ax5.set_title("Euclidean Space")

# Hyperbolic (tree-like)
ax6 = fig.add_subplot(5, 4, 6)
for i in range(3):
    ax6.plot([0, np.cos(i*2*np.pi/3)], [0, np.sin(i*2*np.pi/3)], 'r-')
ax6.set_title("Hyperbolic Tree")

# ---------- Topological Spaces ----------
ax7 = fig.add_subplot(5, 4, 7)
circle = np.linspace(0, 2*np.pi, 100)
ax7.plot(np.cos(circle), np.sin(circle))
ax7.set_title("Topology-Invariant Feature (Circle)")

# ---------- Graph-Based ----------
ax8 = fig.add_subplot(5, 4, 8)
G_nodes = np.array([[0, 0], [1, 1], [2, 0], [1, -1]])
for (i, j) in [(0, 1), (1, 2), (1, 3), (2, 3)]:
    ax8.plot([G_nodes[i,0], G_nodes[j,0]], [G_nodes[i,1], G_nodes[j,1]], 'k-')
ax8.scatter(G_nodes[:,0], G_nodes[:,1], c='orange', s=100)
ax8.set_title("Graph Representation")

# ---------- Latent Spaces ----------
ax9 = fig.add_subplot(5, 4, 9)
data = np.random.randn(200, 10)  # High-dim
from sklearn.decomposition import PCA
latent = PCA(2).fit_transform(data)
ax9.scatter(latent[:, 0], latent[:, 1], alpha=0.6)
ax9.set_title("Latent Embedding Space")

# ---------- Frequency Domain ----------
ax10 = fig.add_subplot(5, 4, 10)
t = np.linspace(0, 1, 400)
signal = np.sin(2*np.pi*5*t) + 0.5*np.sin(2*np.pi*10*t)
freq = np.fft.fft(signal)
ax10.plot(np.abs(freq[:200]))
ax10.set_title("Frequency Domain (FFT)")

# ---------- Algebraic Structures ----------
ax11 = fig.add_subplot(5, 4, 11)
square = np.array([[0,0],[1,0],[1,1],[0,1],[0,0]])
rot = np.array([[0,-1],[1,0]]) @ square.T
ax11.plot(square[:,0], square[:,1], 'b-')
ax11.plot(rot[0], rot[1], 'r-')
ax11.set_title("Group Symmetry (Rotation)")

# ---------- Attention ----------
ax12 = fig.add_subplot(5, 4, 12)
attention = np.random.rand(5, 5)
ax12.imshow(attention, cmap="Blues")
ax12.set_title("Attention Weights")

# ---------- Complex Numbers ----------
ax13 = fig.add_subplot(5, 4, 13)
z = np.exp(1j*np.linspace(0, 2*np.pi, 100))
ax13.plot(z.real, z.imag)
ax13.set_title("Complex Plane")

# ---------- Energy-Based ----------
ax14 = fig.add_subplot(5, 4, 14, projection='3d')
X, Y = np.meshgrid(np.linspace(-2,2,50), np.linspace(-2,2,50))
Z = -(X**2 + Y**2)
ax14.plot_surface(X, Y, Z, cmap="viridis")
ax14.set_title("Energy Landscape")

# ---------- Metric Learning ----------
ax15 = fig.add_subplot(5, 4, 15)
points = np.array([[0,0],[1,1],[2,0]])
ax15.scatter(points[:,0], points[:,1], c=['r','g','b'], s=100)
for i in range(3):
    for j in range(i+1,3):
        d = np.linalg.norm(points[i]-points[j])
        ax15.plot([points[i,0], points[j,0]], [points[i,1], points[j,1]], 'k--')
        ax15.text((points[i,0]+points[j,0])/2, (points[i,1]+points[j,1])/2, f"{d:.2f}")
ax15.set_title("Metric Learning Distances")

# ---------- Density ----------
ax16 = fig.add_subplot(5, 4, 16)
data = np.random.randn(200)
ax16.hist(data, bins=20, density=True, alpha=0.6, color='g')
ax16.plot(x, norm.pdf(x, 0, 1), 'r--')
ax16.set_title("Density Estimation")

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate points on a circle (a topological loop)
theta = np.linspace(0, 2*np.pi, 50)
circle_x = np.cos(theta) + 0.05*np.random.randn(50)  # add small noise
circle_y = np.sin(theta) + 0.05*np.random.randn(50)

# Generate some "noise" points inside the circle
noise_x = 0.3*np.random.randn(20)
noise_y = 0.3*np.random.randn(20)

# Plot
plt.figure(figsize=(6,6))
plt.scatter(circle_x, circle_y, c='b', label='Points on loop')
plt.scatter(noise_x, noise_y, c='g', alpha=0.6, label='Noise points')

# Highlight the "hole" in the middle
circle = plt.Circle((0,0), 0.5, color='r', fill=False, linestyle='--', linewidth=2, label='Topological Hole')
plt.gca().add_patch(circle)

plt.gca().set_aspect('equal')
plt.title("Topological Feature (1D Hole in Data)")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate some complex numbers
z = np.array([1+1j, 2+2j, -1+2j, -2-1j, 1-2j, 2+0.5j])

# Extract real and imaginary parts
real = z.real
imag = z.imag

# Plot complex numbers
plt.figure(figsize=(6,6))
plt.axhline(0, color='black', linewidth=0.8)
plt.axvline(0, color='black', linewidth=0.8)

plt.scatter(real, imag, c='b', s=80, label='Complex numbers')

# Add arrows from origin to each point
for x, y in zip(real, imag):
    plt.arrow(0, 0, x, y, head_width=0.1, head_length=0.1,
              fc='r', ec='r', alpha=0.6)

# Add unit circle (|z|=1)
theta = np.linspace(0, 2*np.pi, 200)
plt.plot(np.cos(theta), np.sin(theta), 'g--', label='Unit Circle')

plt.gca().set_aspect('equal')
plt.xlim(-3, 3)
plt.ylim(-3, 3)
plt.xlabel("Real Axis")
plt.ylabel("Imaginary Axis")
plt.title("Complex Plane Representation")
plt.legend()
plt.show()
```
